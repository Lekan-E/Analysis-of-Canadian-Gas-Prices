```python
# Import libraries
import pandas as pd
```

```python
# Import csv
premium = pd.read_csv(r'./Premium Unleaded Gasoline.csv')
diesel = pd.read_csv(r'./Diesel.csv')
regular_gas = pd.read_csv(r'./Regular Gasoline.csv')
```

```python
# Map provinces with their code
province = {
    'Nunavut':'NU',
    'Ontario' : 'ON',
    'Northwest Territories' : 'NT',
    'Yukon': 'YT',
    'British Columbia':'BC',
    'Alberta':'AB',
    'Saskatchewan':'SK',
    'Manitoba':'MB',
    'Quebec':'QC',
    'New Brunswick':'NB',
    'Nova Scotia':'NS',
    'Prince Edward Island':'PE',
    'Newfoundland and Labrador':'NL'
}

# Convert to dataframe
province_df = pd.DataFrame(province.items())

# Rename columns
province_df.rename(columns={0:'Province', 1:'Code'}, inplace= True)

# Save as csv
#province_df.to_csv(r'./Province Codes.csv', index=False)

# Concat tables 
df = pd.concat([premium, diesel, regular_gas], axis = 0)

# Extract the date
```

#### Assign province codes

```python
# Write a code that assign the province based on the geo 
def province_matching(row):

    global province

    for key in province:
        if key in row['GEO']:
            return key
    
df['Province'] = df.apply(province_matching, axis = 1)

df.head()
```

#### Extract the fuel type from the text

```python
def fueltype(row):
    fuel_type = {
        'Premium unleaded gasoline at self service filling stations':'Premium',
        'Diesel fuel at self service filling stations':'Diesel',
        'Regular unleaded gasoline at self service filling stations':'Regular'
    }

    for key in fuel_type:
        if key in row['Type of fuel']:
            return fuel_type[key]

# Apply fule type function
df['Fuel Type'] = df.apply(fueltype, axis=1)

# Drop unwanted columns
df.drop(columns=['DGUID','UOM_ID','SCALAR_FACTOR','SCALAR_ID', 'Type of fuel',
                 'VECTOR','STATUS', 'SYMBOL','TERMINATED','COORDINATE','DECIMALS'], inplace = True)

# Drop null rows
df = df.dropna()
# Export
df.to_csv(fr'./Combined fuel prices from {df['REF_DATE'].min()} to {df['REF_DATE'].max()}.csv', index=False)
df.head()
```
